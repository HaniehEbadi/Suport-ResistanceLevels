```python
n = 30
r = 2 #range max , min
rD = 200 #Distance of Diagonal line and MAX_MIN
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
df = pd.read_csv('data.csv')
df = pd.DataFrame(df)
df.sort_values(by='date_', inplace = True)
df = df[len(df)-n:]
df = df.reset_index()
df.head()
```

```python
df = df[['date_','close_price']]
df
```

```python
from datetime import datetime
df['ordinalDate'] = df['date_'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d').date().toordinal() )
df
```

```python
prices = df['close_price']

yMINs = []
yMINs.append(np.nan)
for i in range(r-1,len(prices)-r+1):
    for j in range(1,r):
        if prices[i] < prices[i-j] and prices[i] < prices[i+j]:
            if j == r-1:
                yMINs.append(prices[i])
        elif prices[i] > prices[i-j] and prices[i] > prices[i+j]:
            if j == r-1:
                yMINs.append(prices[i])
        else:
            yMINs.append(np.nan)
            break
yMINs.append(np.nan)
df['MIN_MAX'] = yMINs

df2 = df[df['MIN_MAX'].notna()]

df2,df2.index
```

```python
def Eq(x1,x2,y1,y2):
    slope = (y2-y1)/(x2-x1)
    intercept = y1 - slope*x1
    return (slope,intercept)

plt.figure(figsize=(20, 10))
plt.plot(df['ordinalDate'],df['close_price'])
plt.grid()

eq = {}
mList = []
for i in df2.index:
    for j in df2.index:
        m , c = Eq(df2['ordinalDate'][i],df2['ordinalDate'][j],df2['MIN_MAX'][i],df2['MIN_MAX'][j])
        if i != j:
            if m not in mList:
                mList.append(m)
                key = str(i)+str(j)
                eq[key] = [m,c]

# for m in mList:
#     for key in list(eq.keys()):
#         if 0 < abs(eq[key][0] - m) < 10:
#             print(key,eq[key][0],m)
#             del eq[key]
#         elif eq[key][0] == m:
#     mList.remove(m)

for key in eq:
    y = lambda x: eq[key][0]*x + eq[key][1]
    k = 0
    for i in df2.index:
        if abs(y(df2['ordinalDate'][i]) - df2['MIN_MAX'][i]) < rD:
            k += 1
    print(k)
    if k > 5:
        plt.plot(df2['ordinalDate'],y(df2['ordinalDate']))
        for i in df2.index:
            if abs(y(df2['ordinalDate'][i]) - df2['MIN_MAX'][i]) < rD:
                plt.scatter(df2['ordinalDate'][i],y(df2['ordinalDate'][i]))

eq
```
